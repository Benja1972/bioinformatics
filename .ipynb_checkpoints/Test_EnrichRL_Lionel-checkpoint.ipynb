```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import EnrichRLib as erl
```

```python
gn_dt = pd.read_csv('test_from_Lionel/RPFSS_vs_RPFDS_1000_1.5_underexpressed_genes_expression.csv',
                   sep='\t')

gn_dt.head()
```

```python
gl = list(gn_dt['genesymbol'].str.upper())
```

```python
# List of gene sets as above
gss = [ 
       'GO_Biological_Process_2018',
       'GO_Cellular_Component_2018',
       'GO_Molecular_Function_2018',
       'KEGG_2016',
       'Reactome_2016'
       ]
```

```python
enrr = erl.enrich_gs(gl,gss)
```

```python
enrr.head()
```

```python
import seaborn as sns
enrr.sort_values('p-Val', axis=0, inplace = True)
ds = enrr.head(20)

f, ax = plt.subplots()
sns.barplot(y=ds.index,
            x='-log10(p-Val)',
            ax = ax, 
            color="Red", 
            data = ds)
ax.set_title('All terms')
```

```python
# For futher analysis it is convinient to filter terms by p-value for
enrfl = enrr[enrr['p-Val']<0.05]
len(enrfl)
```

```python
dd = erl.dist_matrx(gl,enrfl)
```

```python
## Show cluster based on kappa-score closeness
%matplotlib inline
grid = sns.clustermap(dd, cmap='Blues', figsize=(16, 16))
```

```python
## Cluster: this calculate and add cluster number column
enrfl = erl.cluster(gl,enrfl)

# Make additional calculation on existing columns for visualization
enrfl.loc[:,'ass_genes_percnt'] = 100*enrfl.loc[:,'num_list']/enrfl.loc[:,'num_term']
enrfl.sort_values('cluster', axis=0, inplace = True)

# use consistent discrete palette
cm = 'tab20'

ds = enrfl.head(40)

f, ax = plt.subplots(figsize=(18.5, 8.5))
sns.barplot(y=ds.index,
            x='-log10(p-Val)',
            ax = ax, 
            hue ='cluster',
            dodge=False,
            data = ds,
            palette = cm)
ax.set_title('All terms')
```

```python
f, ax = plt.subplots(figsize=(18.5, 8.5))
sns.barplot(y=ds.index,
            x='ass_genes_percnt',
            ax = ax, 
            hue ='cluster', 
            dodge=False,
            data = ds,
            palette = cm)
ax.set_title('All terms')
ax.set_xlabel('%Genes/Term')
```

```python
## Graph and network
## control conectivity by kappa score parameter, default kappa=0.4


%matplotlib
plt.ioff()
nt, nt_tb, G = erl.make_graph(gl, enrfl, kappa=0.4, draw=True, palette=cm)
plt.show()
```
